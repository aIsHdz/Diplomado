```python
import pandas as pd
```

```python
df = pd.read_csv('/content/salaries.csv')
df.head()
```

```python
df['work_year'] = df.work_year.astype(str)
```

```python
from pandas.api.types import CategoricalDtype

categorias = CategoricalDtype(categories = ['EN', 'MI', 'SE', 'EX'], ordered = True)
df['experience_level'] = df.experience_level.astype(categorias)
df.info()
```

```python
df['same_country'] = df.apply(lambda x:x.employee_residence == x.company_location, axis = 1)
df
```

```python
def format_title(title, subtitle = None, subtitle_font_size=14):
  title = f'{title}'
  if not subtitle:
    return title
  subtitle = f'{subtitle}'
  return f'{title}{subtitle}'
```

```python
fig = px.bar(df1, x = 'experience_level', y = 'Count', text = 'Count')
fig.update_layout(
    xaxis_title = 'Nivel de Experiencia',
    yaxis_title = 'Conteo',
    title_text = format_title('1','2'),
    template = 'gridon'
)
fig
```
